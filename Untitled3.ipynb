```python
import numpy as np
import pandas as pd
from scipy.special import ndtri
import math
import random
```

```python
def makeRandomAverage(num):
    sum = 0;
    for i in range(num):
        sum += random.uniform(0, 40)
    return sum/num

total = 10000
x = np.linspace(1,total,total)
#print(x.tolist())
i1 = sorted(list(map(lambda x: makeRandomAverage(1), x.tolist())))
i11 = sorted(list(map(lambda x: makeRandomAverage(11), x.tolist())))
i101 = sorted(list(map(lambda x: makeRandomAverage(101), x.tolist())))
i1001 = sorted(list(map(lambda x: makeRandomAverage(1001), x.tolist())))
prob = sorted(list(map(lambda x: x/total, x.tolist())))

```

```python
df = pd.DataFrame()
df['x'] = x
df['prob'] = prob
df['i1'] = i1
df['i11'] = i11
df['i101'] = i101
df['i1001'] = i1001
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
plt.figure(num=None, figsize=(14, 8), dpi=180, facecolor='w', edgecolor='k')
plt.plot(ndtri(df.prob), df.i1, 'r--', ndtri(df.prob), df.i11, 'b-', ndtri(df.prob), df.i101, 'g-', ndtri(df.prob), df.i1001, 'y-')
plt.xlabel('Probability')
plt.figure(num=None, figsize=(14, 8), dpi=180, facecolor='w', edgecolor='k')
plt.plot(df.prob, df.i1, 'r--', df.prob, df.i11, 'b-', df.prob, df.i101, 'g-', df.prob, df.i1001, 'y-')
plt.xlabel('Probability')
plt.title('CDF')
a = [1,11,101,1001]
b = [np.var(df.i1),np.var(df.i11),np.var(df.i101),np.var(df.i1001)]
print(np.polyfit(np.log(a),np.log(b),1))
plt.figure(num=None, figsize=(14, 8), dpi=180, facecolor='w', edgecolor='k')
plt.loglog(a,b, basex=2.718, basey=2.718)
plt.xlabel('I Values')
plt.ylabel('Variance')
```

```python
z = pd.read_csv('/Users/markmetzger/Desktop/npl.csv')
porosity = sorted(z.P.tolist())
d = pd.DataFrame()
d['x'] = np.linspace(1,len(porosity),len(porosity))
d['P'] = porosity
prob = sorted(list(map(lambda x: x/len(porosity), d.x.tolist())))
d['prob'] = prob
plt.figure(num=None, figsize=(14, 8), dpi=180, facecolor='w', edgecolor='k')
plt.plot(ndtri(df.prob), df.i1, 'r--', ndtri(df.prob), df.i11, 'b-', ndtri(df.prob), df.i101, 'g-', ndtri(df.prob), df.i1001, 'y-')
plt.plot(ndtri(d.prob), d.P, 'k--')
plt.title('CDF With Porosity Data')

def scrambled(orig):
    dest = orig[:]
    random.shuffle(dest)
    return dest

x = [0]*1000
y = [[]]*1000
for i in range(100):
    copy = scrambled(d.P.tolist())
    for j in range(1000):
        x[j] = j*2+1
        y[j] = y[j][:]
        y[j].append(np.average(copy[0:j*2+1]))

y = list(map(lambda x: np.var(x), y))


plt.figure(num=None, figsize=(14, 8), dpi=180, facecolor='w', edgecolor='k')
plt.loglog(x,y, basex=10, basey=10)
plt.title('Variance vs I Values for Porosity Data (The data is always roughly independent because the fit is a straight line)')
plt.xlabel('I Values')
plt.ylabel('Variance')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def P(r, t):
    return 1000 - (158*100*1/4/3.14159265/50/20)*(np.log(r*r/20/t) + 0.57721)
```

```python
def forward(dt):
    return (P(0.25, 1 + dt) - P(0.25, 1))/dt
def backward(dt):
    return (P(0.25, 1) - P(0.25, 1 - dt))/dt
def center(dt):
    return (P(0.25, 1+dt) - P(0.25, 1 - dt))/(2*dt)

df = pd.DataFrame()
d = np.linspace(0.001, 0.01, 10)
d1 = forward(d)
d2 = backward(d)
d3 = center(d)
df['dt'] = d
df['forward'] = d1
df['backward'] = d2
df['center'] = d3
print(df)
temp = plt.figure()
plt.plot(df.dt, df.forward, 'r', df.dt, df.backward, 'b', df.dt, df.center, 'g')
plt.title('Forward, Backward, and Center Derivative vs dt')
plt.show()
```

```python
d = np.linspace(0.001, 0.01, 10)
df2 = pd.DataFrame()
df2['dt'] = d
df2['forward'] = 1.257324-d1
df2['backward'] = d2-1.257324
print(df2)
temp = plt.figure()
plt.plot(df2.dt, df2.forward, 'r', df2.dt, df2.backward, 'b')
plt.title('Forward and Backward Derivative Error vs dt (E = 0.62dt + 9e-6)')
plt.show()
print(np.polyfit(df2['dt'], df2.forward, 1))
```

```python
d = np.linspace(0.001, 0.01, 10)
df3 = pd.DataFrame()
df3['dt'] = d
df3['dt^2'] = d**2
df3['center'] = d3-1.257324
print(df3)
temp = plt.figure()
plt.plot(df3['dt^2'], df3.center)
plt.title('Center Derivative Error vs dt^2 (E = 0.42dt^2 + 5e-8)')
plt.show()
print(np.polyfit(df3['dt^2'], df3.center, 1))
```

```python
def secondDeriv(dr):
    return (P(.25+dr, 1) - 2*P(0.25,1) + P(.25-dr, 1))/(dr**2)

d = np.linspace(0.0001, 0.001, 10)
df4 = pd.DataFrame()
df4['dr'] = d
df4['derivative'] = secondDeriv(d)
print(df4)
temp = plt.figure()
plt.plot(df4['dr'], df4.derivative)
plt.title('Second Derivative vs dr')
plt.show()
```

```python
d = np.linspace(0.0001, 0.001, 10)
df5 = pd.DataFrame()
df5['dr'] = d
df5['dr^2'] = d**2
df5['error'] = df4['derivative'] - 40.2343696596
print(df5)
temp = plt.figure()
plt.plot(df5['dr^2'], df5.error)
plt.title('Error vs dr^2 (E = 312dr^2 + 2.5e-7)')
plt.show()
print(np.polyfit(df5['dr^2'], df5.error, 1))
```
